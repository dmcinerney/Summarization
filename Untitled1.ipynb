```python
import parameters as p
from summarize import setup, new_model
from data import get_data
from utils import summarize, get_text_triplets
import torch
```

```python
model_file = 'EMNLP/FinalTests/Transformer_generator/checkpoint33/model_state.tpkl'
vectorizer = setup(checkpoint_path=None, device='cuda:0', pointer_gen=False, use_transformer=True, with_coverage=False, model_file=model_file)
data = get_data(p.VAL_FILE, vectorizer, with_oov=False, aspect_file=p.ASPECT_FILE)
```

```python
batch = data[0:1]
```

```python
model = new_model(vectorizer, data.dataset.aspects).eval()
model.load_state_dict(torch.load(p.MODEL_FILE, map_location=p.DEVICE))
```

```python
store = {'mask':torch.ones((1, 400, 1)).cuda().requires_grad_()}
aspect_results = summarize(batch, model, beam_size=4, store=store)
```

```python
store["summary"].keys()
```

```python
attentions = torch.tensor(aspect_results[0][0][4][0], device='cuda:0')
shape = store["summary"]["text_states"].shape
grads = torch.zeros((shape[1], attentions.shape[0], shape[2]))
for i in range(attentions.shape[0]):
    context_vector = store["summary"]["text_states"][0].transpose(0,1) @ attentions[i]
    for j in range(shape[2]):
        context_vector[j].backward(retain_graph=True)
        grads[:,i,j] = store["summary"]["mask"].grad[0,:,0]
        store["mask"].grad = store["mask"].grad*0
    print(i)
```

```python
grads_mag = (grads**2).sum(-1)**(1/2)
grads_mag
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(50, 40))
im = ax.imshow(grads_mag.transpose(0,1))
plt.savefig("TransformerPointerGeneratorGradients")
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(50, 40))
im = ax.imshow(attentions.cpu())
plt.savefig("TransformerPointerGeneratorAttentions")
```

```python
attention_traceback = torch.tensor(attentions).cpu()
for l in [1,0]:
    self_attention = sum(torch.eye(400)+store["summary"]["encoder_transformer_attns"][0,l,:,i,:].cpu().detach()
                         for i in range(8))
    attention_traceback = (self_attention.transpose(0,1) @ attention_traceback.transpose(0,1)).transpose(0,1)
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(50, 40))
im = ax.imshow(attention_traceback.cpu())
plt.savefig("TransformerPointerGeneratorAttentionTracebacks")
```
